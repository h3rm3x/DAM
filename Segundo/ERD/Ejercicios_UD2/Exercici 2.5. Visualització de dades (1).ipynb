1. Importa llibreries i dataset
Importa seaborn com a sns, matplotlib.pyplot com a plt i aplica %matplotlib inline

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

2. Boxplot múltiple per comparar espècies

Representa sepal_length, sepal_width, petal_length, petal_width en 4 subgràfics.

```python
# Reuse available seaborn/plt and irisdf; create 4 subplots for the four measurements
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
irisdf = sns.load_dataset('iris')
for ax, feat in zip(axes.flatten(), features):
    sns.boxplot(data=irisdf, x='species', y=feat, ax=ax)
    ax.set_title(feat.replace('_', ' ').capitalize())
    ax.set_xlabel('')
    ax.tick_params(axis='x', rotation=45)

fig.suptitle('Boxplots de mesures florals per espècie', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

3. Pairplot amb regressió

```python
# Pairplot with regression lines between numerical features, separated by species
sns.pairplot(irisdf, hue='species', diag_kind='kde', kind='reg', 
              plot_kws={'scatter_kws': {'s':20, 'alpha':0.6}, 'line_kws': {'linewidth':1}})
plt.suptitle('Pairplot with regression per species', y=1.02)
plt.show()
```

4. Heatmap de correlacions

```python
import numpy as np
num_cols = irisdf.select_dtypes(include=[np.number])
corr = num_cols.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Mapa de correlació de les variables numèriques (Iris)')
plt.show()
```

6. Crea un PairGrid personalitzat amb diferents tipus de gràfics

Utilitzarem Seaborn PairGrid, més flexible que pairplot.

```python
pairgrid = sns.PairGrid(data=irisdf, hue='species')
pairgrid.map_diag(sns.kdeplot)
pairgrid.map_upper(sns.scatterplot, s=20)
pairgrid.map_lower(sns.kdeplot, fill=True)

for ax_row in pairgrid.axes:
    for ax in ax_row:
        ax.relim()
        ax.autoscale()
```

7. Crea un FacetGrid per visualitzar relacions segons espècie

```python
facetgrid = sns.FacetGrid(irisdf, col='species')
facetgrid.map(sns.histplot, 'sepal_length', edgecolor='black').set_ylabels('Count')
```

```python
facetgrid = sns.FacetGrid(irisdf, col='species')
facetgrid.map(sns.histplot, 'sepal_length', bins=15, kde=True)
facetgrid.set_axis_labels('Sepal Length', 'Count')
```
