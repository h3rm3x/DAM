```python
# 1. Importació i càrrega de dades (1 punt)

# Importa pandas com a pd i carrega els tres fitxers CSV (clients.csv, products.csv, orders.csv).
# Mostra les 5 primeres files de cadascun dels tres dataframes.
```

```python
import pandas as pd
import numpy as np

# Carregar els fitxers CSV
clients_df = pd.read_csv("data/clients.csv")
products_df = pd.read_csv("data/products.csv")
orders_df = pd.read_csv("data/orders.csv")

# Mostrar les primeres 5 files
print("Clients:")
display(clients_df.head())

print("Productes:")
display(products_df.head())

print("Comandes:")
display(orders_df.head())
```

```python
# 2. Neteja bàsica (1 punt)

# Mostra si existeixen valors nuls o duplicats en cadascun dels dataframes.

# Elimina'ls, si n'hi ha, fent servir les funcions adequades (dropna, drop_duplicates).

# Mostra el nombre de files abans i després del procés.
```

```python
# Comprovació de valors nuls
print("Valors nuls per dataframe:")
print("Clients:\n", clients_df.isnull().sum())
print("Productes:\n", products_df.isnull().sum())
print("Comandes:\n", orders_df.isnull().sum())

# Comprovació de duplicats
print("\nDuplicats:")
print("Clients:", clients_df.duplicated().sum())
print("Productes:", products_df.duplicated().sum())
print("Comandes:", orders_df.duplicated().sum())

# Eliminació de possibles duplicats o valors nuls
clients_df = clients_df.dropna().drop_duplicates()
products_df = products_df.dropna().drop_duplicates()
orders_df = orders_df.dropna().drop_duplicates()

print("\nMides finals després de la neteja:")
print("Clients:", clients_df.shape)
print("Productes:", products_df.shape)
print("Comandes:", orders_df.shape)
```

```python
# 3. Fusió de datasets (1 punt)

# Uneix els tres datasets per obtenir un únic dataframe sales_df amb tota la informació.
# Empra les claus següents:

# client_id per unir clients_df amb orders_df

# product_id per unir amb products_df

# Mostra les primeres 5 files de sales_df.
```

```python
# Fusió de clients amb comandes
sales_df = pd.merge(orders_df, clients_df, on="client_id", how="left")

# Fusió amb productes
sales_df = pd.merge(sales_df, products_df, on="product_id", how="left")

display(sales_df.head())
```

```python
# 4. Creació de columnes noves (1 punt)

# Afegeix una columna total_price que sigui el resultat de units * price.

# Converteix order_date a tipus datetime.

# Afegeix les columnes year i month extretes d’aquesta data.
```

```python
# Columna de preu total
sales_df["total_price"] = sales_df["units"] * sales_df["price"]

# Conversió de data
sales_df["order_date"] = pd.to_datetime(sales_df["order_date"])

# Extracció de l’any i el mes
sales_df["year"] = sales_df["order_date"].dt.year
sales_df["month"] = sales_df["order_date"].dt.month

display(sales_df.head())
```

```python
# 5. Agrupacions i càlculs estadístics (1,5 punts)

# Calcula el total de vendes (total_price) per categoria.

# Calcula la mitjana d’edat dels clients per ciutat.

# Mostra la mitjana de preu i nombre total d’unitats venudes per categoria.
```

```python
# 1. Total de vendes per categoria
vendes_per_categoria = sales_df.groupby("category")["total_price"].sum()
print("Total de vendes per categoria:")
display(vendes_per_categoria)

# 2. Mitjana d’edat per ciutat
edat_per_ciutat = sales_df.groupby("city")["age"].mean()
print("\nMitjana d’edat per ciutat:")
display(edat_per_ciutat)

# 3. Mitjana de preu i total d’unitats per categoria
resum_categoria = sales_df.groupby("category").agg({
    "price": "mean",
    "units": "sum"
})
print("\nResum per categoria:")
display(resum_categoria)
```

```python
# 6. Filtratge i ordenació (1 punt)

# Mostra les vendes del 2023 per a la categoria electronics.

# Ordena el resultat pel total_price descendent.
```

```python
# Filtrar vendes de 2023 per electronics
vendes_2023 = sales_df[
    (sales_df["year"] == 2023) & (sales_df["category"] == "electronics")
]

# Ordenar pel total de vendes descendent
vendes_2023 = vendes_2023.sort_values("total_price", ascending=False)

display(vendes_2023)
```

```python
# 7. Anàlisi de benefici (1,5 punts)

# Afegeix una columna cost_unitari, que sigui un valor aleatori entre el 60% i el 90% del preu (price).

# Calcula una nova columna benefici = (price - cost_unitari) * units.

# Agrupa per categoria i mostra:

#  total de vendes

#  benefici total

#  marge mitjà en percentatge → (benefici / total_price) * 100
```

```python
# Generar un cost unitari aleatori entre el 60% i el 90% del preu
np.random.seed(42)
sales_df["cost_unitari"] = sales_df["price"] * np.random.uniform(0.6, 0.9, len(sales_df))

# Calcular el benefici
sales_df["benefici"] = (sales_df["price"] - sales_df["cost_unitari"]) * sales_df["units"]

# Agrupar per categoria
resum_benefici = sales_df.groupby("category").agg({
    "total_price": "sum",
    "benefici": "sum"
})
resum_benefici["marge_percent"] = (resum_benefici["benefici"] / resum_benefici["total_price"]) * 100

print("Resum de benefici per categoria:")
display(resum_benefici)
```

```python
# 8. Normalització per grups (1 punt)

# Normalitza la columna benefici dins de cada categoria, dividint-la pel màxim benefici d’aquell grup.

# Afegeix el resultat en una nova columna benefici_normalitzat.
```

```python
# Normalitzar el benefici dins de cada categoria
sales_df["benefici_normalitzat"] = sales_df.groupby("category")["benefici"].transform(lambda x: x / x.max())

display(sales_df[["category", "benefici", "benefici_normalitzat"]].head())
```

```python
# 9. Filtratge avançat i exportació (1 punt)

# Mostra només les categories on la mitjana del benefici normalitzat sigui superior a 0.5.

# Desa aquest subconjunt en un fitxer CSV anomenat filtered_sales.csv.
```

```python
# Calcular la mitjana del benefici normalitzat per categoria
mitjana_benefici = sales_df.groupby("category")["benefici_normalitzat"].mean()

# Filtrar categories amb mitjana > 0.5
categories_filtrades = mitjana_benefici[mitjana_benefici > 0.5].index

# Subconjunt filtrat
filtered_sales = sales_df[sales_df["category"].isin(categories_filtrades)]

display(filtered_sales)

# Exportar a CSV
filtered_sales.to_csv("filtered_sales.csv", index=False)
print("Fitxer 'filtered_sales.csv' desat correctament.")
```
