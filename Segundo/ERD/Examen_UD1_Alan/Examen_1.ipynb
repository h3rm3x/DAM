```python
# 1. Importació i càrrega de dades (1 punt)

# Importa pandas com a pd i carrega els tres fitxers CSV (clients.csv, products.csv, orders.csv).
# Mostra les 5 primeres files de cadascun dels tres dataframes.
```

```python
import pandas as pd
clients_df = pd.read_csv('./data/clients.csv', header=0)
products_df = pd.read_csv('./data/products.csv', header=0)
orders_df = pd.read_csv('./data/orders.csv', header=0)
```

```python
# 2. Neteja bàsica (1 punt)

# Mostra si existeixen valors nuls o duplicats en cadascun dels dataframes.

# Elimina'ls, si n'hi ha, fent servir les funcions adequades (dropna, drop_duplicates).

# Mostra el nombre de files abans i després del procés.
```

```python
for df, name in zip([clients_df, products_df, orders_df], ['clients_df', 'products_df', 'orders_df']):
    print(f"DataFrame: {name}")
    print("Antes de la limpieza:")
    print(f"Valores nulos:\n{df.isna().sum()}")
    print(f"Filas duplicadas: {df.duplicated().sum()}")

    initial_rows = df.shape[0]
    df.dropna(inplace=True)
    df.drop_duplicates(inplace=True)
    final_rows = df.shape[0]

    print("Después de la limpieza:")
    print(f"Filas antes: {initial_rows}, Filas después: {final_rows}\n")
```

```python
# 3. Fusió de datasets (1 punt)

# Uneix els tres datasets per obtenir un únic dataframe sales_df amb tota la informació.
# Empra les claus següents:

# client_id per unir clients_df amb orders_df

# product_id per unir amb products_df

# Mostra les primeres 5 files de sales_df.
```

```python
sales_df = clients_df.join(orders_df.set_index('client_id'), on='client_id').join(products_df.set_index('product_id'), on='product_id') #join de las tres tablas
#eliminamos las filas con valores nulos en sales_df
sales_df.dropna(inplace=True)
sales_df.head()
```

```python
# 4. Creació de columnes noves (1 punt)

# Afegeix una columna total_price que sigui el resultat de units * price.

# Converteix order_date a tipus datetime.

# Afegeix les columnes year i month extretes d’aquesta data.
```

```python
sales_df['total_price'] = sales_df['units'] * sales_df['price'] # creamos la columna total_price
sales_df['order_date'] = pd.to_datetime(sales_df['order_date']) # cambiamos el tipo de order_date a datetime
sales_df['year'] = sales_df['order_date'].dt.year #creamos la columna year extrayendo el año de order_date
sales_df['month'] = sales_df['order_date'].dt.month #creamos la columna month extrayendo el mes de order_date
sales_df.head()
```

```python
# 5. Agrupacions i càlculs estadístics (1,5 punts)

# Calcula el total de vendes (total_price) per categoria.

# Calcula la mitjana d’edat dels clients per ciutat.

# Mostra la mitjana de preu i nombre total d’unitats venudes per categoria.
```

```python
print("Total de ventas por categoría:")
sales_by_category = sales_df.groupby('category')['total_price'].sum() 
print(sales_by_category)
print("\nEdad media de los clientes por ciudad:")
avg_age_by_city = sales_df.groupby('city')['age'].mean()
print(avg_age_by_city)
print("\nMedia de precio y unidades vendidas por categoría:")
category_stats = sales_df.groupby('category').agg({'price': 'mean', 'units': 'sum'})
print(category_stats)
```

```python
# 6. Filtratge i ordenació (1 punt)

# Mostra les vendes del 2023 per a la categoria electronics.

# Ordena el resultat pel total_price descendent.
```

```python
sales_df_2023_electronics = sales_df[(sales_df['year'] == 2023) & (sales_df['category'] == 'electronics')] # filtramos las ventas del 2023 para la categoria electronics
sales_df_2023_electronics.sort_values(by='total_price', ascending=False, inplace=True) # ordenamos por total_price descendente
sales_df_2023_electronics.head() #mostramos las primeras 5 filas del resultado
```

```python
# 7. Anàlisi de benefici (1,5 punts)

# Afegeix una columna cost_unitari, que sigui un valor aleatori entre el 60% i el 90% del preu (price).

# Calcula una nova columna benefici = (price - cost_unitari) * units.

# Agrupa per categoria i mostra:

#  total de vendes

#  benefici total

#  marge mitjà en percentatge → (benefici / total_price) * 100
```

```python
import numpy as np
sales_df['cost_unitari'] = sales_df['price'] * np.random.uniform(0.6, 0.9, size=sales_df.shape[0]) # creamos la columna cost_unitari con valores aleatorios entre el 60% y 90% del precio
sales_df['benefici'] = (sales_df['price'] - sales_df['cost_unitari']) * sales_df['units'] # creamos la columna benefici
print("Primeras 5 filas de sales_df con las nuevas columnas:")
print(sales_df[['price', 'cost_unitari','units', 'benefici']].head())
category_analysis = sales_df.groupby('category').agg( #agrupamos por categoria y calculamos las métricas solicitadas
    total_vendes=('total_price', 'sum'),
    benefici_total=('benefici', 'sum')
)

category_analysis['marge_mitja_percentatge'] = ((category_analysis['benefici_total'] / category_analysis['total_vendes']) * 100 ) 
print("Análisis de ventas por categoría:")
category_analysis
```

```python
# 8. Normalització per grups (1 punt)

# Normalitza la columna benefici dins de cada categoria, dividint-la pel màxim benefici d’aquell grup.

# Afegeix el resultat en una nova columna benefici_normalitzat.
```

```python
group_max = sales_df.groupby('category')['benefici'].transform('max')
sales_df['benefici_normalitzat'] = np.where(group_max > 0, sales_df['benefici'] / group_max, 0)
print("Columna 'benefici_normalitzat' creada. Primeras filas:")
sales_df.head()
```

```python
# 9. Filtratge avançat i exportació (1 punt)

# Mostra només les categories on la mitjana del benefici normalitzat sigui superior a 0.5.

# Desa aquest subconjunt en un fitxer CSV anomenat filtered_sales.csv.
```

```python
category_df = sales_df.groupby('category')['benefici_normalitzat'].mean().reset_index()
filtered_categories = category_df[category_df['benefici_normalitzat'] > 0.5]['category']
filtered_sales_df = sales_df[sales_df['category'].isin(filtered_categories)]
filtered_sales_df.to_csv('filtered_sales.csv', index=False)
print("DataFrame filtrado guardado en 'filtered_sales.csv'. Primeras filas:")
print(filtered_sales_df.head())

print("El csv esta vacio ya que ningun categoria tiene una media de benefici normalitzat superior a 0.5")
```
