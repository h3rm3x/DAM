Importa pandas com a pd

```python
import pandas as pd
import numpy as np
```

Crea un nou dataframe anomenat user_df. Aquest carregarà les dades dels usuaris del fitxer u.user que trobareu al directori data.

Els noms de les columnes han d'esser 'user_id', 'age', 'gender', 'occupation' i 'zipcode'. El delimitador emprat és la barra vertical '|'. La columna de l'índex serà la 0 (user_id)

Finalment mostra el dataframe escrivint user_df a una nova linia

```python
# import user data 
user_data = pd.read_csv('data/u.user', sep='|', header=None, names=[ 'user_id', 'age', 'gender', 'occupation', 'zip_code' ])
user_data.set_index('user_id')
user_df = pd.DataFrame({ 'user_id' : user_data['user_id'], 'age' : user_data['age'], 'gender' : user_data['gender'], 'occupation' : user_data['occupation'], 'zip_code' : user_data['zip_code'] })
print(user_df[0:])
```

Emprant l'slicing vist als materials de numpy, mostra les rows de la 5 a la 10.

```python
print(user_df.loc[5:10]) #imprimir de la fila 5 a la 10
```

Mostra el dataset sols amb les columnes age, gender i occupation. Desa-ho a un nou dataset anomenat anon_user_df

```python
anon_user_df = pd.DataFrame(user_df.drop(columns=['zip_code'], axis=1)) #eliminar columnes
print(anon_user_df) #imprimir les columnes
```

A anon_user_df, filtra les rows sols a les que tenguin la ocupació 'student'

```python
print(anon_user_df[anon_user_df['occupation'] == 'student']) #imprimir els estudiants
```

Agrupa els usuaris anonimitzats per ocupació. Desa el DataFrameGroupBy com a user_group.

```python
user_grouped = anon_user_df.groupby('occupation')
```

Descriu user_group emprant la funció describe()

```python
user_grouped.describe()
```

Mostra per cada ocupació, quants usuaris hi ha (empra qualsevol de les columnes per a contar)

```python
user_grouped.size() #comptar el nombre d'usuaris per ocupació
```

Mostra per cada ocupació, la mitjana d'edat. Ordena els resultats per edat descendent

```python
user_grouped = user_grouped['age'].mean()
user_grouped = user_grouped.sort_values(ascending=False) #agrupació per ocupació
print(user_grouped)
```

Importa les dades de u.data. Anomena'l rating_df. Les columnes seran user_id, movie_id, rating i timestamp. Canvia el delimitador al que correspongui.

```python
rating_df = pd.read_csv('data/u.data', sep='\t', header=None, names=['user_id', 'movie_id', 'rating', 'timestamp'])
print(rating_df) #imprimir el dataframe
```

Crea un nou dataframe anomenat user_rating_df. Fes que sigui el resultat de la join entre rating_df i user_df emprant la columna user_id.

```python
user_rating_df = pd.merge(user_df, rating_df, on='user_id')
print(user_rating_df) #imprimir el nou dataframe
```

Esborra les files que tenguin algún NaN. Aplica a la funció l'atribut inplace=True.

```python
# eliminar las filas con NaN en cualquier columna
user_rating_df.dropna(inplace=True)
user_rating_df
```

Reanomena la columna timestamp a datetime emprant la funció rename i fent el canvi persistent (inplace=True)

```python
user_rating_df.rename(columns= { 'timestamp': 'datetime' }, inplace=True) #renombrar columna
print(user_rating_df) #imprimir el nou dataframe
```

Importa from datetime, datetime. Aplicant una funció lamda, converteix el format de timestamp a datetime.

```python
from datetime import datetime
user_rating_df['datetime'] = pd.to_datetime(user_rating_df['datetime'], unit='s') #convertir a datetime
print(user_rating_df) #imprimir el nou dataframe
```
