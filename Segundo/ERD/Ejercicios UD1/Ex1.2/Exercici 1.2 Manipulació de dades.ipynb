# Exercici

Escriu una funció que rebi un diccionari amb les notes dels alumnes d’un curs i retorni una sèrie amb la nota mínima, la màxima, la mitjana i la desviació típica.

```python
import pandas as pd
import numpy as np


# Ejemplo de uso
notas = pd.read_csv('./notas_alumnos.csv', header=0, names=['ID_alumno', 'matematicas', 'lengua', 'ciencias'], sep=',')
notas.set_index('ID_alumno', inplace=True)
notas
```

```python
def calcular_estadisticas(notas=notas):

    estadisticas = pd.DataFrame(columns=['Media Mates', 'Desviacion_Estandar Mates', 'Maximo Mates', 'Minimo Mates',
                                       'Media Lengua', 'Desviacion_Estandar Lengua', 'Maximo Lengua', 'Minimo Lengua',
                                       'Media Ciencias', 'Desviacion_Estandar Ciencias', 'Maximo Ciencias', 'Minimo Ciencias'],
                               index=['Valor'])

    estadisticas['Media Mates']['Valor'] = notas['matematicas'].mean()
    estadisticas['Desviacion_Estandar Mates']['Valor'] = notas['matematicas'].std()
    estadisticas['Maximo Mates']['Valor'] = notas['matematicas'].max()
    estadisticas['Minimo Mates']['Valor'] = notas['matematicas'].min()

    estadisticas['Media Lengua']['Valor'] = notas['lengua'].mean()
    estadisticas['Desviacion_Estandar Lengua']['Valor'] = notas['lengua'].std()
    estadisticas['Maximo Lengua']['Valor'] = notas['lengua'].max()
    estadisticas['Minimo Lengua']['Valor'] = notas['lengua'].min()

    estadisticas['Media Ciencias']['Valor'] = notas['ciencias'].mean()
    estadisticas['Desviacion_Estandar Ciencias']['Valor'] = notas['ciencias'].std()
    estadisticas['Maximo Ciencias']['Valor'] = notas['ciencias'].max()
    estadisticas['Minimo Ciencias']['Valor'] = notas['ciencias'].min()

    return estadisticas
calcular_estadisticas()
```
