# Exercici

Escriu una funció que rebi un diccionari amb les notes dels alumnes d’un curs i retorni una sèrie amb la nota mínima, la màxima, la mitjana i la desviació típica.

```python
import pandas as pd
import numpy as np


# Ejemplo de uso
notas = pd.read_csv('./notas_alumnos.csv', header=0, names=['ID_alumno', 'matematicas', 'lengua', 'ciencias'], sep=',')
notas.set_index('ID_alumno', inplace=True)
notas
```

```python
def calcular_estadisticas(notas=notas):

    estadisticas = pd.DataFrame(columns=['Media Mates', 'Desviacion_Estandar Mates', 'Maximo Mates', 'Minimo Mates',
                                       'Media Lengua', 'Desviacion_Estandar Lengua', 'Maximo Lengua', 'Minimo Lengua',
                                       'Media Ciencias', 'Desviacion_Estandar Ciencias', 'Maximo Ciencias', 'Minimo Ciencias'],
                               index=['Valor'])

    estadisticas.at['Valor', 'Media Mates'] = notas['matematicas'].mean()
    estadisticas.at['Valor', 'Desviacion_Estandar Mates'] = notas['matematicas'].std()
    estadisticas.at['Valor', 'Maximo Mates'] = notas['matematicas'].max()
    estadisticas.at['Valor', 'Minimo Mates'] = notas['matematicas'].min()

    estadisticas.at['Valor', 'Media Lengua'] = notas['lengua'].mean()
    estadisticas.at['Valor', 'Desviacion_Estandar Lengua'] = notas['lengua'].std()
    estadisticas.at['Valor', 'Maximo Lengua'] = notas['lengua'].max()
    estadisticas.at['Valor', 'Minimo Lengua'] = notas['lengua'].min()

    estadisticas.at['Valor', 'Media Ciencias'] = notas['ciencias'].mean()
    estadisticas.at['Valor', 'Desviacion_Estandar Ciencias'] = notas['ciencias'].std()
    estadisticas.at['Valor', 'Maximo Ciencias'] = notas['ciencias'].max()
    estadisticas.at['Valor', 'Minimo Ciencias'] = notas['ciencias'].min()

    return estadisticas
calcular_estadisticas()
```
