Importa els rating des de /data/rating.data. Anomena'l rating_df.

Els noms de les columnes han d'esser 'user_id','movie_id','rating' i 'timestamp'. Canvia el delimitador al que correspongui.

Solució

```python
rating_df = pd.read_csv('rating.data' , sep='	', names=[ 'user_id', 'movie_id', 'rating', 'timestamp' ], header=0)
rating_df.set_index('user_id', inplace=True)
rating_df
```

Crea un nou dataframe anomenat movie_rating_df. Fes que sigui el resultat de la join entre rating_df i movie_df emprant la columna movie_id.

Solució

```python
movie_rating_df = pd.merge(movie_df, rating_df, on='movie_id')
movie_rating_df.set_index('movie_id', inplace=True)
movie_rating_df
```

Eliminar filas con NaN

Solució

```python
movie_rating_df.dropna(inplace=True)
movie_rating_df
```

enombrar columna timestamp com rating_time

Solució

```python
movie_rating_df.rename(columns={'timestamp': 'rating_time'}, inplace=True)
movie_rating_df
```

Convertir rating_time a datetime

Solució

```python
movie_rating_df.rename(columns={'rating_time': 'datetime'}, inplace=True)
movie_rating_df
```

Importa from datetime, datetime. Aplicant una funció lamda, converteix el format de timestamp a datetime.

Solució

```python
from datetime import datetime
movie_rating_df['datetime'] = movie_rating_df['datetime'].apply(lambda x: datetime.fromtimestamp(x))

```
